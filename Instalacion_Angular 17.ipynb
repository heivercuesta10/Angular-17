# **Angular 17**



Construye aplicaciones web sofisticadas y dinámicas con Angular, uno de los frameworks que está revolucionando el mundo. Aprenderás a crear aplicaciones con componentes reutilizables, por medio de las herramientas de desarrollo y testing que ofrece Angular.







Angular Está pensado para dividir un proyecto en componentes y ser reutilizadas en proyectos medianos y grandes.

# **Competidores de Angular**

Sus principales competidores son Vue (proyecto iniciado por Evan You en 2014) y React ( proyecto iniciado por Facebook en 2013)

Instalación de Angular CLI
Para instalar este software lo hacemos desde la misma línea de comandos de Node.js (por eso lo instalamos primero), debemos ejecutar el siguiente comando:

npm install -g @angular/cli

Instalación de Angular CLI




Es importante el -g para que se instale en forma global.
